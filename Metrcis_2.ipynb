```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
```

```python
classif = pd.read_csv('classification.csv')
```

```python
classif
TP = classif[(classif.true == classif.pred)&(classif.true == 1)].shape[0]
FP = classif[(classif.true == 0)&(classif.pred == 1)].shape[0]
FN = classif[(classif.true == 1)&(classif.pred == 0)].shape[0]
TN = classif[(classif.true == classif.pred)&(classif.true == 0)].shape[0]
```

```python
print(TP, FP, FN, TN)
```

```python
print( 
    accuracy_score(classif.true,classif.pred),
    precision_score(classif.true,classif.pred),
    recall_score(classif.true,classif.pred),
    f1_score(classif.true,classif.pred)
        )
```
