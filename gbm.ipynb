```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from typing import List, Tuple
import matplotlib.pyplot as plt
```

```python
gbm_data = pd.read_csv('gbm-data.csv')
```

```python
gbm_data.head()
```

```python
X = gbm_data.drop('Activity', axis = 1)
y = gbm_data['Activity']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                      test_size=0.8, 
                                      random_state=241)
```

```python
def sigmoid(y_pred: np.array) -> np.array:
    return 1.0 / (1.0 + np.exp(-y_pred))

def plot(test_loss, train_loss):
    plt.figure()
    plt.plot(test_loss, 'r', linewidth=2)
    plt.plot(train_loss, 'g', linewidth=2)
    plt.legend(['test', 'train'])
    plt.show()

learn_rate = [1, 0.5, 0.3, 0.2, 0.1]
min_loss_results = {}
clf = GradientBoostingClassifier( n_estimators=250, verbose=True, random_state=241, learning_rate = 1)
for lr in learn_rate:
    clf = GradientBoostingClassifier( n_estimators=250, verbose=True, random_state=241, learning_rate = lr)
    clf.fit(X_train,y_train)
    loss_train = [log_loss(y_train,sigmoid(y_pred)) for y_pred in clf.staged_decision_function(X_train)]
    loss_test = [log_loss(y_test,sigmoid(y_pred)) for y_pred in clf.staged_decision_function(X_test)]
    plot(loss_test, loss_train)
    
    min_loss_value = min(loss_test)
    min_loss_index = loss_test.index(min_loss_value)
    min_loss_results[lr] = min_loss_value, min_loss_index
    
    print("Min loss " + str(round(min_loss_value, 5)) + " at n_estimators= " + str(min_loss_index))
```

```python
model = RandomForestClassifier(n_estimators=min_loss_index, random_state=241)
model.fit(X_train, y_train)

y_pred = model.predict_proba(X_test)[:, 1]
test_loss = log_loss(y_test, y_pred)

print(3, f"{test_loss:.2f}")
```
